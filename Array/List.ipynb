```python
import ctypes
```

```python
class List:
    def __init__(self):
        self.size = 1
        self.n = 0
        self.A = self.__make_array(self.size)

    def __make_array(self, capacity):
        return (capacity*ctypes.py_object)()

    def __resize(self, new_capacity):
        B = self.__make_array(new_capacity)
        self.size = new_capacity

        for i in range(self.n):
            B[i] = self.A[i]

        self.A = B

    def __len__(self):
        return self.n

    def __repr__(self):
        res = ''
        for i in range(self.n):
            res = res + str(self.A[i]) + ', '
        res = res[:-2]
        return '[' + res + ']'

    def __getitem__(self, index):
        if isinstance(index, int):
            if index >= self.n:
                raise IndexError("Index out of bounds")
            elif index >= 0:
                return self.A[index]
            elif -1 >= index >= -len(self):
                index = self.n + index
                return self[index]
            else:
                raise IndexError("Index out of bounds")
        elif isinstance(index, slice):
            other = List()
            for i in range(index.start, index.stop, index.step):
                other.append(self.A[i])
            return other

    def __delitem__(self, index):
        if index >= self.n:
            raise IndexError("Index out of bounds")
        elif index >= 0:
            for i in range(index, self.n - 1):
                self.A[i] = self.A[i+1]
            self.n -= 1
        elif -1 >= index >= -len(self):
            index = self.n + index
            del self[index]
        else:
            raise IndexError("Index out of bounds")

    def __add__(self, other):
        another = List()
        for i in range(self.n):
            another.append(self.A[i])
        for i in range(other.n):
            another.append(other.A[i])
        return another

    def __mul__(self, times):
        other = List()
        for _ in range(times):
            for i in range(self.n):
                other.append(self.A[i])
        return other

    def __max__(self):
        max = self.A[0]
        for i in range(self.n):
            if max < self.A[i]:
                max = self.A[i]
        return max

    def __min__(self):
        min = self.A[0]
        for i in range(self.n):
            if min > self.A[i]:
                min = self.A[i]
        return min

    def __sum__(self):
        sum = 0
        for i in range(self.n):
            sum += self.A[i]
        return sum

    def append(self, element):
        if self.n == self.size:
            self.__resize(self.size * 2)
        self.A[self.n] = element
        self.n += 1

    def insert(self, index, value):
        if self.n == self.size:
            self.__resize(self.size*2)
        for i in range(self.n, index, -1):
            self.A[i] = self.A[i-1]
        self.A[index] = value
        self.n += 1

    def extend(self, other):
        for i in range(other.n):
            self.append(other.A[i])

    def pop(self):
        if self.n == 0:
            raise IndexError("pop from empty list")
        self.n -= 1
        return self.A[self.n]

    def clear(self):
        self.n = 0
        self.size = 0

    def find(self, value):
        for i in range(self.n):
            if self.A[i] == value:
                return i
        raise ValueError("value not in list")

    def remove(self, value):
        index = self.find(value)
        del self[index]

    def sort(self):
        # Insertion sort
        if self.n <= 1:
            return

        for i in range(1, self.n):
            key = self.A[i]
            j = i-1

            while j >= 0 and key < self.A[j]:
                self.A[j+1] = self.A[j]
                j -= 1

            self.A[j+1] = key
```
