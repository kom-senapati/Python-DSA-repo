```python
class Node:
    def __init__(self, value=None) -> None:
        self.data = value
        self.next = None
```

```python
class LinkedList:
    def __init__(self) -> None:
        self.head = None
        self.n = 0

    def __len__(self):
        return self.n

    def __repr__(self) -> str:
        res = ""
        curr = self.head
        while curr != None:
            res += f"{curr.data} -> "
            curr = curr.next
        return res[:-4]

    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self.n += 1

    def append(self, value):
        new_node = Node(value)

        if self.n == 0:
            self.head = new_node
            self.n += 1
            return

        curr = self.head
        while curr.next != None:
            curr = curr.next

        curr.next = new_node
        self.n += 1

    def insert(self, index, value):
        n = 0
        curr = self.head
        while n != index - 1:
            curr = curr.next
            n += 1

        new_node = Node(value)
        new_node.next = curr.next
        curr.next = new_node
        self.n += 1

    def insert_after(self, after, value):
        new_node = Node(value)
        curr = self.head
        while curr.data != after:
            curr = curr.next
            if curr == None:
                return "Item not found"

        new_node.next = curr.next
        curr.next = new_node
        self.n += 1

    def clear(self):
        self.head = None
        self.n = 0

    def shift(self):
        if self.n == 0:
            return "Empty linkedin list"
        self.head = self.head.next
        self.n -= 1

    def pop(self):
        if self.n == 0:
            return "Empty linkedin list"

        elif self.n == 1:
            data = self.head.data
            self.head = None
            self.n = 0
            return data

        curr = self.head
        while curr.next.next != None:
            curr = curr.next

        data = curr.next.data
        curr.next = None
        self.n -= 1

        return data

    def remove(self, value):
        if self.n == 0:
            return "Empty linkedin list"
        elif self.n == 1:
            if self.head.data == value:
                self.pop()
            return

        if self.head.data == value:
            return self.shift()

        curr = self.head
        while curr.next.data != value:
            curr = curr.next
            if curr == None:
                return "Value not found"

        curr = curr.next.next
        self.n -= 1

    def search(self, value):
        curr = self.head
        n = 0
        while curr != None:
            if curr.data == value:
                return n
            curr = curr.next
            n += 1

        return -1

    def __getitem__(self, index):
        if index < 0:
            index += self.n
        curr = self.head
        n = 0
        while curr != None:
            if n == index:
                return curr.data
            curr = curr.next
            n += 1

        return "Indexerror"

    # Problem 1
    def max(self):
        max = self.head.data
        curr = self.head
        while curr:
            if max < curr.data:
                max = curr.data
            curr = curr.next

        return max

    def replace_max(self, value):
        max = self.max()
        curr = self.head
        while curr:
            if curr.data == max:
                curr.data = value
            curr = curr.next

    # Problem 2
    def sum_odd(self):
        if self.n < 2:
            return 0

        curr = self.head.next
        sum = 0
        while curr:
            sum += curr.data
            if not curr.next:
                break
            curr = curr.next.next

        return sum

    # Problem 3
    def reverse(self):
        back = None
        curr = self.head
        while curr:
            temp = curr.next
            curr.next = back
            back = curr
            curr = temp

        self.head = back

    # Problem 4
    def change_sent(self):
        curr = self.head
        while curr:
            if curr.data in "*/":
                if curr.next.data in "*/":
                    curr.next = curr.next.next
                    curr.next.data = curr.next.data.upper()
                curr.data = " "
            curr = curr.next

    def trverse_str(self):
        curr = self.head
        while curr:
            print(curr.data, end="")
            curr = curr.next
```

```python
numbers = LinkedList()
for i in range(6):
    numbers.append(i)
print(numbers)
print("sum =", numbers.sum_odd())
```

```python
print(numbers)
numbers.replace_max(100)
print(numbers)
```

```python
numbers.reverse()
print(numbers)
```

```python
sentence = LinkedList()
for ch in "The//sky*is*/blue.":
    sentence.append(ch)

sentence.trverse_str()
sentence.change_sent()
print("\n")
sentence.trverse_str()
```
